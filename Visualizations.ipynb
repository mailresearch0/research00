```python
import datetime
import numpy as np
import csv
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib as mpl
import matplotlib.ticker as ticker
import matplotlib.cm as cm
import seaborn as sns
import matplotlib.colors as colors
import matplotlib.gridspec as gridspec # spaces between graphs

%matplotlib inline

# Read file 1 for net score
df1 = pd.read_csv("Data_Note7_1.csv", header=0)
# Select only the columns that matters for the score per day
df1[['date', 'text', 'score']]
# summarize the score per day in a dataframe
score_day=df1.groupby('date',as_index=False)[['score']].sum()
#change the dates to dates
score_day['date']=pd.to_datetime(score_day['date'],format = '%Y/%m/%d')

# Read file 2 for heatmap
df2 = pd.read_csv("f_score_day_Note7.csv", header=1, index_col=0 )

# Read file 3 for p-chart
df3 = pd.read_csv("P_Chart_Note7.csv", header=0)
# Select only the columns that matters for p-chart
data=pd.DataFrame(df3,columns=['Date','p_bar','p','Out','UCL','LCL'])
data['Date']=pd.to_datetime(data['Date'],format = '%d/%m/%Y')

# First Graph
fig = plt.figure(figsize=(20,15)) # Change the graph size
x1=plt.subplot(3,1,1)

plt.subplots_adjust(wspace=0.05, hspace=0.05) # Space between graphs
#fig.set_size_inches(w,h)
plt.plot(score_day['date'],score_day['score'])
plt.xticks(rotation = 30, fontsize = 12)
plt.grid(color='w', linestyle='-') 
plt.grid(False)
plt.margins(x=0)
plt.yticks(fontsize=12)
plt.xticks([])
plt.ylabel('Net Score per day', fontsize = 12)
plt.title('Galaxy Note7', fontsize = 16)

# Second graph 
x2=plt.subplot(3,1,2)
x2 = sns.heatmap(df2, norm=colors.PowerNorm(gamma=1./2.), 
                cmap = 'viridis_r', cbar=False) 
plt.ylabel('Frequency Score', fontsize = 12)
plt.yticks(fontsize = 12)
sns.set_style("whitegrid")
sns.despine()
plt.xticks(rotation = 30, fontsize = 12)
plt.xticks([])
#plt.xticks()

# Third graph
x3=plt.subplot(3,1,3)
plt.plot(data['Date'], data['p_bar'], 'bo', color ='b', alpha= 0.5,label='Neg p', ms=5)
plt.step(data['Date'], data['UCL'], color ='grey', label = 'UCL',alpha = 0.5)
plt.plot(data['Date'], data['LCL'], '-', color ='grey', label = 'LCL',alpha = 0.5)
plt.plot(data['Date'], data['Out'], 'bo', color = 'r', alpha = 0.7, ms=5)
plt.plot(data['Date'], data['p'], '--', color ='g', alpha = 0.5, label = 'Global mean')
plt.xticks(rotation = 30, fontsize = 12)
plt.margins(x=0)
#plt.xticks([])
plt.grid(False)
#sns.despine()
plt.yticks(fontsize=12)
plt.xlabel('Date', fontsize = 14)
plt.ylabel('Proportion', fontsize = 12)
#plt.set_facecolor('#ffffff')
plt.show()
```
